```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gws_dataset=pd.read_csv("merged_gw.csv")
gws_dataset.columns
```

```python
fwd_gws_dataset= gws_dataset.loc[gws_dataset.position == 'FWD']
mid_gws_dataset= gws_dataset.loc[gws_dataset.position == 'MID']
def_gws_dataset= gws_dataset.loc[gws_dataset.position == 'DEF']
goal_gws_dataset= gws_dataset.loc[gws_dataset.position == 'GK']
```

```python
plt.xlabel('ict_index')
plt.ylabel('total_points')
plt.scatter(fwd_gws_dataset.ict_index , fwd_gws_dataset.total_points , color='red' , marker='+')
```

```python
from mpl_toolkits import mplot3d
plt.style.use('seaborn-poster')
```

```python
fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection='3d')
```

```python
fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection='3d')
ax.grid()
ax.plot3D(fwd_gws_dataset.transfers_in, fwd_gws_dataset.transfers_out, fwd_gws_dataset.total_points)
ax.set_title('3D Parametric Plot')
# Set axes label
ax.set_xlabel('transfers_in', labelpad=20)
ax.set_ylabel('transfers_out', labelpad=20)
ax.set_zlabel('total_points', labelpad=20)

plt.show()
```
